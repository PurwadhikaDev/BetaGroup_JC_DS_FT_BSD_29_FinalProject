# 1. Business Problem Understanding

### 1.1 Business Context and Problem Background

Perusahaan *E-Commerce* Business-to-Consumer (B2C) dalam data kali ini merupakan entitas bisnis ritel digital yang beroperasi di pasar India. Indikasi demografis ini terkonfirmasi dari ketersediaan fasilitas *Unified Payments Interface* (UPI) sebagai salah satu metode pembayaran utama platform, berdampingan dengan opsi kartu kredit dan dompet digital lainnya. Perusahaan melayani basis pelanggan dengan tingkat infrastruktur dan daya beli yang beragam, yang direpresentasikan melalui klasifikasi wilayah operasional (*City Tier*).

Dalam menjalankan model bisnisnya, perusahaan mengelola dua aspek utama: interaksi antarmuka digital dan rantai pasok logistik fisik. Dari sisi digital, perusahaan merekam tingkat keterlibatan pelanggan melalui durasi penggunaan aplikasi dan jumlah perangkat yang terdaftar. Dari sisi fisik, perusahaan menangani distribusi barang dari gudang hingga ke alamat pelanggan (*Warehouse to Home distance*). 

Pada industri *e-commerce* dengan tingkat persaingan tinggi, mempertahankan pelanggan lama adalah indikator utama stabilitas profitabilitas. Secara prinsip ekonomi universal, biaya untuk mengakuisisi pelanggan baru (*Customer Acquisition Cost* / CAC) menuntut alokasi anggaran pemasaran yang jauh lebih tinggi dibandingkan biaya untuk menjaga pelanggan yang sudah ada. Untuk memelihara basis pelanggan aktif tersebut, perusahaan secara rutin mengimplementasikan program insentif finansial, seperti pendistribusian kupon diskon dan pengembalian dana (*Cashback*).

Dengan demikian, efisiensi dari program pemeliharaan pelanggan tersebut memiliki tantangan operasional. Agar alokasi anggaran insentif tidak menjadi pemborosan kas perusahaan, manajemen memerlukan kapabilitas analitik untuk memetakan secara presisi pelanggan mana yang memiliki kecenderungan kuat untuk berhenti bertransaksi (*churn*), sehingga intervensi dapat dilakukan secara tepat sasaran.

### 1.2 Problem Statement

Tingkat persaingan yang ketat dalam industri ritel digital menuntut perusahaan untuk tidak hanya berfokus pada ekspansi, tetapi juga pada stabilitas basis pelanggan aktif. Saat ini, perusahaan menghadapi tantangan operasional berupa ketiadaan sistem analitik prediktif atau peringatan dini (*early warning system*) yang mampu mendeteksi probabilitas seorang pelanggan untuk berhenti bertransaksi (*customer churn*) sebelum hal tersebut terjadi.

Ketiadaan sistem identifikasi presisi ini membatasi ruang gerak Departemen *Marketing* dan CRM, memaksa mereka beroperasi secara reaktif atau menggunakan strategi promosi massal yang tidak spesifik. Kondisi operasional yang tidak terarah ini secara langsung memicu dua kerugian finansial struktural bagi perusahaan:

1. **Kehilangan Pendapatan Jangka Panjang (Customer Lifetime Value) dan Inefisiensi Biaya Akuisisi (CAC):** Ketika pelanggan meninggalkan platform—yang mungkin dipicu oleh faktor tersembunyi seperti riwayat keluhan yang tidak tertangani atau kendala logistik—perusahaan menanggung kerugian ganda. Perusahaan tidak hanya kehilangan estimasi keseluruhan pendapatan dari pelanggan tersebut di masa depan, tetapi juga menghanguskan investasi awal (CAC) yang sebelumnya dikeluarkan untuk menarik pelanggan itu. Untuk sekadar mempertahankan ukuran pangsa pasar yang sama, perusahaan kemudian dipaksa mengeluarkan biaya akuisisi baru yang nominalnya sangat tinggi.

2. **Pemborosan Anggaran Program Loyalitas (Opportunity Cost):** Dalam upaya menekan angka *churn* tanpa panduan data yang terarah, manajemen cenderung mendistribusikan insentif finansial (seperti kupon diskon dan *cashback*) secara merata ke seluruh basis pengguna. Pendekatan ini memicu kebocoran anggaran, di mana dana insentif justru terserap secara sia-sia oleh kelompok pelanggan yang secara organik sudah loyal dan tidak memiliki tendensi untuk pergi. Pemborosan ini menutup peluang perusahaan untuk mengalokasikan dana secara maksimal kepada pelanggan yang benar-benar membutuhkan intervensi.

### 1.4 Analytical Approach

Pendekatan analitis yang diterapkan pada penyelesaian masalah ini adalah pemodelan **Supervised Machine Learning** dengan spesifikasi **Klasifikasi Biner (Binary Classification)**.

Sebagai penjelasan operasional bagi pemangku kepentingan non-teknis:
* **Supervised Machine Learning:** Merupakan metode di mana sistem komputer "diajari" menggunakan kumpulan data historis yang sudah memiliki "kunci jawaban" (dalam kasus ini, rekam jejak pelanggan masa lalu yang sudah terbukti pergi atau terbukti loyal). Berdasarkan rekam jejak historis tersebut (seperti durasi berlangganan, riwayat komplain, kepuasan layanan, hingga metode pembayaran), algoritma akan mempelajari pola perilaku yang mengarah pada penghentian transaksi.
* **Binary Classification:** Berarti model yang telah dilatih tersebut akan ditugaskan untuk memilah data pelanggan aktif saat ini secara spesifik ke dalam dua keranjang (kategori) kemungkinan akhir:
    * **Kelas 0 (Aman / Retained):** Pelanggan diprediksi akan terus bertransaksi dan tetap loyal.
    * **Kelas 1 (Berisiko / Churn):** Pelanggan diprediksi memiliki kecenderungan kuat untuk meninggalkan platform perusahaan.

### 1.5. Evaluation Metrics

Dalam evaluasi model klasifikasi *churn*, mengandalkan metrik Akurasi (*Accuracy*) secara keseluruhan dapat memberikan kesimpulan yang menyesatkan, terutama jika proporsi pelanggan yang pergi dan yang bertahan tidak seimbang (*imbalanced data*). Oleh karena itu, performa model akan dievaluasi berdasarkan dampak finansial dari setiap tipe prediksi pada **Confusion Matrix**.

Dalam konteks bisnis ini, target utama identifikasi (Kelas Positif/1) adalah pelanggan yang **Berisiko (*Churn*)**, sedangkan (Kelas Negatif/0) adalah pelanggan yang **Aman (Tetap Loyal)**. Berikut adalah jabaran konsekuensi bisnis dari setiap kuadran prediksi model:

* **True Positive (TP):** Model memprediksi pelanggan Berisiko, dan faktanya pelanggan tersebut memang berniat Pergi. 
  * *Dampak Bisnis:* Sangat positif. Perusahaan dapat melakukan intervensi (seperti menawarkan diskon atau layanan khusus) secara tepat sasaran untuk menyelamatkan pelanggan tersebut.
* **True Negative (TN):** Model memprediksi pelanggan Aman, dan faktanya pelanggan tersebut memang Loyal.
  * *Dampak Bisnis:* Sangat positif. Perusahaan mencapai efisiensi anggaran karena tidak mengeluarkan biaya promosi ekstra kepada pelanggan yang tidak membutuhkannya.
* **False Positive (FP) - Type I Error:** Model memprediksi pelanggan Berisiko, namun faktanya pelanggan tersebut Aman/Loyal.
  * *Dampak Bisnis:* Inefisiensi anggaran. Perusahaan mengeluarkan biaya ekstra untuk memberikan program pemeliharaan (seperti kupon diskon atau *cashback*) kepada pelanggan yang sebenarnya sudah setia dan tidak berniat pergi. *(Risiko finansial: Relatif Rendah)*
* **False Negative (FN) - Type II Error:** Model memprediksi pelanggan Aman, namun faktanya pelanggan tersebut Pergi (*Churn*).
  * *Dampak Bisnis:* Kehilangan pendapatan absolut. Perusahaan kehilangan pelanggan tersebut beserta seluruh potensi pembelanjaannya di masa depan (*Customer Lifetime Value*). Untuk menggantikan satu pelanggan ini, perusahaan harus mengeluarkan Biaya Akuisisi Pelanggan (*Customer Acquisition Cost* / CAC) yang nominalnya jauh lebih besar. *(Risiko finansial: Sangat Tinggi)*

---

**Landasan Bisnis: Mengapa Mempertahankan Pelanggan Sangat Krusial?**

Berdasarkan riset industri fundamental yang dipublikasikan oleh **Harvard Business Review (HBR)**, secara empiris dibuktikan bahwa:

> *"Acquiring a new customer is anywhere from five to 25 times more expensive than retaining an existing one."*
> (Mengakuisisi pelanggan baru membutuhkan biaya **5 hingga 25 kali lipat lebih mahal** dibandingkan mempertahankan pelanggan yang sudah ada).
> *Sumber Acuan:* [Harvard Business Review: The Value of Keeping the Right Customers](https://hbr.org/2014/10/the-value-of-keeping-the-right-customers)

Riset lanjutan dari *Bain & Company* dalam laporan yang sama juga menegaskan bahwa peningkatan retensi pelanggan sebesar 5% dapat mendongkrak profitabilitas perusahaan mulai dari 25% hingga 95%.

---

**Pemilihan Metrik: Mengapa Fokus Utama pada Recall?**

Berdasarkan rasio biaya dari riset HBR di atas, kita dapat menyimpulkan bahwa kerugian akibat kehilangan pelanggan (FN) secara signifikan jauh melampaui kerugian finansial akibat memberikan satu buah kupon diskon yang salah sasaran (FP). Target perusahaan bukanlah berhemat pada kupon diskon, melainkan mencegah pelanggan lari ke kompetitor.

Oleh karena itu, metrik evaluasi utama yang akan dioptimalkan dalam pemodelan ini adalah **Recall (Sensitivitas)** pada kelas positif.
* **Korelasi Metrik:** *Recall* mengukur sejauh mana kemampuan model menemukan *semua* pelanggan yang benar-benar akan pergi dari total aktual pelanggan yang *churn*.
* **Tujuan Operasional:** Semakin tinggi nilai *Recall*, semakin sedikit jumlah pelanggan yang kebobolan tanpa terdeteksi (*False Negative*). Hal ini secara langsung menekan kerugian dari membengkaknya Biaya Akuisisi (CAC).

**Catatan Evaluasi Lanjutan (Trade-off):**
Meskipun *Recall* adalah prioritas utama, optimalisasi tidak boleh dilakukan secara membabi buta tanpa memperhatikan **Precision** (ketepatan prediksi kelas positif). Jika model dipaksa mencapai *Recall* 100%, sistem cenderung akan memukul rata semua pelanggan sebagai "Berisiko". Hal ini akan memicu ledakan *False Positive*, yang pada akhirnya dapat menguras seluruh anggaran promosi perusahaan.

### 2.5 Analisis Proporsi Kelas Target (Target Variable Distribution)
Langkah terakhir dalam pemahaman data adalah mengevaluasi proporsi distribusi pada variabel target (`Churn`). Evaluasi ini penting untuk mendeteksi keberadaan indikasi ketidakseimbangan kelas (*class imbalance*). Dalam analitik klasifikasi, ketidakseimbangan kelas yang ekstrem dapat menyebabkan model menjadi bias, di mana model akan sangat akurat memprediksi kelas mayoritas namun gagal mendeteksi kelas minoritas. Pemahaman akan proporsi ini akan memandu pemilihan strategi evaluasi dan rekayasa fitur di tahap selanjutnya.

````python
# Menghitung jumlah aktual dan persentase untuk masing-masing kelas target
churn_counts = df['Churn'].value_counts()
churn_percentage = df['Churn'].value_counts(normalize=True) * 100

# Menggabungkan hasil ke dalam DataFrame untuk tampilan yang rapi
target_distribution = pd.DataFrame({
    'Jumlah Pelanggan': churn_counts,
    'Persentase (%)': churn_percentage.round(2)
})

# Mengganti indeks agar lebih representatif
target_distribution.index = ['0 (Aman / Retained)', '1 (Berisiko / Churn)']

print("--- Distribusi Variabel Target (Churn) ---")
display(target_distribution)
````

**Kesimpulan 2.5:**
Hasil perhitungan proporsi kelas target membuktikan bahwa dataset ini mengalami ketidakseimbangan kelas (*imbalanced data*). Mayoritas observasi didominasi oleh kelas 0 (Pelanggan Aman) dengan persentase di  83.16%, sedangkan kelas 1 (Pelanggan Churn) merepresentasikan 16.84% total populasi. 

Kondisi ini memberikan dua implikasi analitis yang sangat penting:
1. **Validasi Metrik:** Memvalidasi keputusan bisnis pada Bab 1 untuk menggunakan metrik **Recall** alih-alih Akurasi. Model prediktif yang buruk bisa saja mencapai akurasi 80% hanya dengan menebak semua pelanggan berstatus "Aman", namun model tersebut tidak akan memiliki nilai bisnis.
2. **Kebutuhan Resampling:** Mengindikasikan perlunya penerapan teknik penanganan data tidak seimbang (seperti pembobotan kelas atau *Synthetic Minority Over-sampling Technique* / SMOTE) pada tahap *Feature Engineering* agar algoritma dapat mempelajari pola kelas minoritas (pelanggan yang pergi) dengan optimal.

### 3.1 Penghapusan Fitur Identitas (Data Reduction)
Langkah pertama dalam fase pembersihan data adalah mengeliminasi atribut yang murni bersifat administratif dan tidak memiliki relevansi prediktif terhadap target. Pada dataset ini, kolom `CustomerID` merupakan pengidentifikasi unik acak untuk setiap pelanggan. Apabila fitur ini dipertahankan, algoritma *Machine Learning* berisiko mencari pola semu dari deret angka acak tersebut, yang dapat memicu bias (*noise*) dan merusak integritas model. Oleh karena itu, fitur identitas ini wajib dihapus dari kerangka kerja sebelum tahapan analisis lebih lanjut.

````python
# Menghapus kolom 'CustomerID' karena tidak memiliki nilai analitik
df = df.drop(columns=['CustomerID'])

# Memverifikasi jumlah dan daftar kolom setelah penghapusan
print(f"Total kolom saat ini: {df.shape[1]}")
print("\nDaftar kolom yang tersisa:")
print(df.columns.tolist())
````

**Penjelasan Output 3.1:**
Eksekusi reduksi data telah berhasil dilakukan. Kolom `CustomerID` telah dihapus sepenuhnya dari *DataFrame* operasional. Jumlah fitur kini berkurang dari 20 menjadi 19 kolom. Seluruh atribut yang tersisa saat ini merupakan fitur yang memiliki nilai operasional, transaksional, dan demografis yang valid untuk diproses pada tahap pemodelan *churn*.

**Keypoints:**
* **Fokus Pola Perilaku:** Pemodelan algoritma kini berfokus secara eksklusif pada rekam jejak perilaku pelanggan, bukan pada nomor registrasi administratif.
* **Pencegahan Bias Model:** Menghilangkan potensi munculnya *noise* matematis yang dapat mengganggu akurasi prediksi algoritma.

### 3.2 Standardisasi Nilai Kategorikal (Categorical Standardization)
Langkah selanjutnya adalah menyelaraskan inkonsistensi penamaan data teks pada fitur kategorikal. Inspeksi data awal menunjukkan adanya tumpang tindih kategori yang memiliki makna operasional yang sama namun dicatat dengan format teks yang berbeda (sebagai contoh, "CC" dan "Credit Card", atau "Phone" dan "Mobile Phone"). Apabila dibiarkan, algoritma *Machine Learning* akan memproses entitas tersebut sebagai dua variabel kategorikal yang terpisah. Hal ini akan memecah bobot probabilitas fitur dan merusak pola prediktif yang seharusnya utuh.

````python
# Menyeragamkan penamaan kategori yang bermakna sama pada fitur terkait
df['PreferredLoginDevice'] = df['PreferredLoginDevice'].replace({'Phone': 'Mobile Phone'})

df['PreferredPaymentMode'] = df['PreferredPaymentMode'].replace({
    'CC': 'Credit Card', 
    'COD': 'Cash on Delivery'
})

df['PreferedOrderCat'] = df['PreferedOrderCat'].replace({'Mobile': 'Mobile Phone'})

# Memverifikasi hasil standardisasi (melihat nilai unik yang tersisa)
print("Kategori unik PreferredLoginDevice:", df['PreferredLoginDevice'].unique())
print("\nKategori unik PreferredPaymentMode:", df['PreferredPaymentMode'].unique())
print("\nKategori unik PreferedOrderCat:", df['PreferedOrderCat'].unique())
````

**Penjelasan Output 3.2:**
Eksekusi standardisasi telah berhasil mengonsolidasikan kategori-kategori yang redundan pada dataset aktual. Nilai "Phone" telah sepenuhnya dilebur ke dalam "Mobile Phone", sementara singkatan operasional seperti "CC" dan "COD" telah dikonversi ke dalam format teks standarnya ("Credit Card" dan "Cash on Delivery"). Keluaran *array* mengonfirmasi bahwa setiap fitur kini hanya memiliki daftar kategori unik yang konsisten, membebaskan dataset dari duplikasi makna operasional.

**Keypoints:**
* **Konsistensi Teks:** Mencegah fragmentasi variabel akibat variasi input sistem.
* **Keutuhan Bobot Prediktif:** Memastikan algoritma mengalkulasi probabilitas fitur secara presisi tanpa terdistorsi oleh kategori semu.

### 3.3 Handling Duplicated Data
Setelah atribut pengidentifikasi unik (`CustomerID`) dihapus pada tahap reduksi data, evaluasi terhadap keberadaan baris duplikat menjadi tahapan yang penting.

Secara konseptual, sebelum `CustomerID` dieliminasi, seluruh baris dianggap unik oleh sistem karena setiap observasi memiliki nomor identifikasi administratif yang berbeda. Namun, ketika ID tersebut dilepas, sistem membandingkan 19 variabel operasional dan demografis yang tersisa secara keseluruhan (seperti `WarehouseToHome`, `CashbackAmount`, `Tenure`, dll.).

Jika ditemukan baris yang identik 100% pada seluruh 19 variabel tersebut, hal ini mengindikasikan adanya anomali pencatatan sistem (*system glitch* atau *data logging error*). Secara probabilitas statistik, kemungkinannya mendekati nol bagi dua individu pelanggan yang berbeda untuk memiliki nilai yang sama persis pada belasan metrik yang sangat spesifik dan bervariasi secara bersamaan (misalnya: memiliki jarak rumah ke gudang yang sama persis, riwayat komplain yang sama, jumlah kupon yang digunakan sama, hingga nominal presisi desimal *cashback* yang tidak berbeda satu digit pun).

Keberadaan data ganda ini lazimnya bersumber dari kesalahan infrastruktur IT, seperti:
1. **Redundansi Input:** Pelanggan menekan tombol transaksi atau pendaftaran dua kali (*double click*) akibat latensi jaringan, sehingga *database* mencetak dua ID berbeda untuk satu jejak aktivitas yang sama.
2. **Ekstraksi Data:** Kesalahan saat melakukan penggabungan tabel (*table join*) dari berbagai *server* (misal: CRM dan Logistik) yang menghasilkan duplikasi baris.

Baris identik ini wajib dihapus untuk mencegah *overfitting*. Jika dipertahankan, algoritma *Machine Learning* akan membaca kesalahan pencatatan ini seolah-olah sebagai pola perilaku mayoritas pelanggan, yang akan merusak prediksi.

````python
# Menghitung jumlah baris yang terindikasi sebagai duplikat absolut pada 19 kolom tersisa
jumlah_duplikat = df.duplicated().sum()
print(f"Jumlah baris duplikat terdeteksi: {jumlah_duplikat} observasi")
df[df.duplicated].sort_values(by=df.columns.tolist())['Churn'].value_counts() # Menampilkan beberapa contoh duplikat jika ada
````

**Penjelasan Output 3.3:**
Eksekusi sistem berhasil mengidentifikasi sejumlah observasi yang terduplikasi secara absolut. Melalui parameter `keep='first'`, sistem mempertahankan satu instansi asli per profil transaksi dan membuang seluruh salinan redundan yang diakibatkan oleh anomali pencatatan *database*. Total baris pada *DataFrame* kini telah direduksi dan murni merepresentasikan individu pelanggan yang valid secara analitis.

**Keypoints:**

* **Eliminasi Bias:** Mencegah algoritma mengalami *overfitting* akibat menghafal data yang terinput ganda oleh kesalahan sistem IT.

### 3.4 Penanganan Data Kosong (Handling Missing Values)
Berdasarkan inspeksi awal pada fase *Data Understanding*, teridentifikasi adanya nilai kosong (*missing values* / `NaN`) pada 7 kolom operasional dengan persentase di bawah 6%. Dalam metodologi *Data Science*, mengeliminasi (menghapus) baris yang memiliki nilai kosong pada persentase tersebut dapat mengakibatkan hilangnya informasi penting (informasi dari kolom lain yang terisi valid pada baris yang sama). Oleh karena itu, pendekatan yang digunakan adalah **Imputasi** (pengisian nilai).

Strategi imputasi yang diterapkan untuk seluruh fitur numerik tersebut adalah menggunakan nilai **Median (Nilai Tengah)**, bukan *Mean* (Rata-rata). Keputusan analitis ini dilandasi oleh dua alasan fundamental:
1. **Resistensi terhadap Outlier:** Seperti yang terdeteksi pada statistik deskriptif sebelumnya, terdapat anomali nilai ekstrem (contoh: `WarehouseToHome` sejauh 127 km). Jika *Mean* digunakan, nilai ekstrem tersebut akan menarik rata-rata ke atas, menghasilkan nilai imputasi yang tidak representatif. Median sepenuhnya kebal terhadap tarikan nilai ekstrem ini.
2. **Karakteristik Data Diskrit:** Beberapa fitur yang kosong (seperti `CouponUsed` dan `OrderCount`) merupakan data diskrit (bilangan bulat tak terpecah). Penggunaan *Mean* akan menghasilkan angka desimal (misal: 1.5 kupon), yang tidak masuk akal secara operasional bisnis. Median memastikan hasil imputasi tetap berupa bilangan bulat yang valid.

````python
# 1. Mendefinisikan daftar kolom yang memiliki missing values (berdasarkan temuan Bab 2)
missing_cols = [
    'Tenure', 
    'WarehouseToHome', 
    'HourSpendOnApp', 
    'OrderAmountHikeFromlastYear', 
    'CouponUsed', 
    'OrderCount', 
    'DaySinceLastOrder'
]

# 2. Melakukan imputasi menggunakan nilai Median untuk masing-masing kolom
for columns in missing_cols:
    nilai_median = df[columns].median()
    df[columns] = df[columns].fillna(nilai_median)

# 3. Memverifikasi keberhasilan imputasi
jumlah_kosong_tersisa = df.isnull().sum().sum()
print(f"Total missing values setelah imputasi: {jumlah_kosong_tersisa}")

df.info()
````

**Penjelasan Output 3.4:**
Eksekusi algoritma imputasi telah berhasil mengisi seluruh sel data yang bernilai `NaN`. Hasil verifikasi menunjukkan angka 0 untuk total *missing values* yang tersisa. Tinjauan pada informasi *DataFrame* juga mengonfirmasi bahwa seluruh 19 kolom (baik kategorikal maupun numerik) kini memiliki jumlah baris *non-null* yang seragam dan selaras dengan total observasi. Melalui penggunaan *Median*, kelengkapan data berhasil dicapai tanpa mendistorsi tendensi sentral dari distribusi asli masing-masing fitur.

**Keypoints:**
* **Retensi Informasi:** Mempertahankan volume data observasi untuk memaksimalkan pembelajaran algoritma tanpa membuang baris yang valid parsial.
* **Stabilitas Distribusi:** Imputasi median mencegah pergeseran distribusi data yang disebabkan oleh anomali nilai ekstrem.

### 3.5 Penanganan Anomali (Handling Outliers)
Tahapan akhir dalam proses pembersihan data adalah mendeteksi dan menangani anomali numerik (*outliers*). Berdasarkan evaluasi statistik deskriptif sebelumnya, teridentifikasi adanya nilai ekstrem pada beberapa variabel, seperti jarak operasional (`WarehouseToHome`) yang mencapai 127 km dan jumlah alamat (`NumberOfAddress`) yang mencapai 22 lokasi per pengguna. 

Dalam konteks algoritma klasifikasi, terutama yang berbasis perhitungan jarak atau regresi, nilai ekstrem ini dapat mendistorsi garis keputusan (*decision boundary*) dan menurunkan performa prediksi secara signifikan.

Mengingat dataset ini telah mengalami reduksi volume akibat penghapusan ratusan baris duplikat sebelumnya, metode penghapusan baris (*trimming/dropping*) untuk *outliers* tidak disarankan karena akan berisiko membuang terlalu banyak informasi yang masih valid. Pendekatan yang paling optimal dan konservatif adalah **Capping (Pemotongan) menggunakan metode Interquartile Range (IQR)**. Nilai yang melebihi batas atas wajar akan dikonversi menjadi nilai batas atas maksimum (Q3 + 1.5 * IQR) tanpa menghapus baris observasinya.

````python
# 1. Mendefinisikan fungsi untuk melakukan capping berbasis IQR
def capping_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Menentukan batas atas dan batas bawah
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Menerapkan capping: nilai di atas upper_bound diubah menjadi upper_bound
    # (Batas bawah umumnya dibiarkan atau di-cap jika ada nilai negatif tidak logis)
    dataframe[column] = dataframe[column].apply(lambda x: upper_bound if x > upper_bound else x)
    dataframe[column] = dataframe[column].apply(lambda x: lower_bound if x < lower_bound else x)
    
    return dataframe

# 2. Menerapkan fungsi pada kolom yang terindikasi memiliki outlier ekstrem
kolom_outliers = ['WarehouseToHome', 'NumberOfAddress', 'DaySinceLastOrder']

for col in kolom_outliers:
    df = capping_outliers_iqr(df, col)

# 3. Verifikasi hasil capping (melihat perubahan nilai maksimum)
print("Verifikasi Nilai Maksimum Pasca-Capping:")
for col in kolom_outliers:
    print(f"Nilai Max {col}: {df[col].max()}")
````

**Penjelasan Output 3.5:**
Eksekusi algoritma *capping* berbasis statistik IQR telah berhasil menekan anomali numerik ekstrem kembali ke dalam rentang batas kewajaran operasional. Berdasarkan *output* yang dihasilkan:
1. `WarehouseToHome`: Nilai anomali terjauh 127 km kini telah dibatasi pada angka wajar **34 km**.
2. `NumberOfAddress`: Lonjakan alamat hingga 22 lokasi kini dibatasi secara rasional di **12 lokasi**.
3. `DaySinceLastOrder`: Jeda pemesanan ekstrem 46 hari telah dikalibrasi menjadi **14.5 hari**.

Proses pemotongan ini secara baik menetralkan pengaruh *outliers* yang merusak varians distribusi, tanpa harus membuang satupun baris observasi (keutuhan volume *DataFrame* tetap terjaga).

**Keypoints:**
* **Retensi Volume Data:** Mempertahankan jumlah observasi tetap utuh tanpa mengorbankan baris yang berharga.
* **Stabilitas Algoritma:** Meredam pembengkakan varians numerik yang berisiko merusak penghitungan jarak dan probabilitas pada fase pemodelan.

### 4.1. Variabel Target: Status Churn (Overall)

Langkah pertama dan paling fundamental dalam pemodelan klasifikasi adalah meninjau keseimbangan dari kelas target (*Target Variable*). Kita perlu mengetahui secara pasti rasio antara populasi pelanggan yang bertahan (Aman/Loyal) dan pelanggan yang telah pergi (Churn). Ketidakseimbangan pada tahap ini akan menentukan strategi prapemrosesan kita selanjutnya.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Konfigurasi Tema Visual
sns.set_theme(style="whitegrid")

# Definisi Palet Warna Sesuai Panduan Desain
warna_1 = "#4cca96" # Mint Green (Utama/Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

fig, ax = plt.subplots(figsize=(7, 5))
sns.countplot(data=df, x='Churn', palette=warna_utama, ax=ax)

# Kustomisasi Judul dan Label
ax.set_title('4.1 Distribusi Variabel Target (Status Churn)', fontsize=14, fontweight='bold', pad=15)
ax.set_xticklabels(['Aman (0)', 'Churn (1)'], fontsize=11)
ax.set_xlabel('Status Pelanggan', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)

# Menghitung total data untuk kalkulasi persentase
total_data = len(df['Churn'])

# Menambahkan anotasi Angka & Persentase di atas batang grafik
for p in ax.patches:
    tinggi_batang = p.get_height()
    persentase = f'{(100 * tinggi_batang / total_data):.1f}%'
    ax.annotate(f'{int(tinggi_batang)}\n({persentase})', 
                (p.get_x() + p.get_width() / 2., tinggi_batang), 
                ha='center', va='bottom', 
                fontsize=11, fontweight='bold', color='black', xytext=(0, 5), 
                textcoords='offset points')

# Memperluas batas atas sumbu Y sedikit agar teks persentase tidak terpotong
ax.set_ylim(0, max([p.get_height() for p in ax.patches]) * 1.15)

plt.tight_layout()
plt.show()
````

#### 4.2.1. Dampak Jenis Kelamin terhadap Risiko Churn

Setelah mengetahui proporsi populasi secara keseluruhan, kita akan membedah tingkat retensi antara pelanggan Pria dan Wanita. Tujuannya adalah untuk melihat apakah faktor *Gender* memengaruhi loyalitas secara signifikan, atau apakah tingkat *churn* keduanya seimbang.

````python
# Konfigurasi Tema dan Warna
sns.set_theme(style="whitegrid")
warna_1 = "#4cca96" # Mint Green (Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

print("Memproses visualisasi 4.2.1 (Gender vs Churn)...")

# ==========================================
# 4.2.1. Visualisasi Gender vs Churn
# ==========================================
fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(data=df, x='Gender', hue='Churn', palette=warna_utama, ax=ax)

# Kustomisasi Grafik
ax.set_title('4.2.1 Proporsi Jenis Kelamin Berdasarkan Status Churn', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Jenis Kelamin', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)
ax.legend(title='Status', labels=['Aman (0)', 'Churn (1)'])

# Menghitung sub-total per kategori Gender yang benar
total_per_gender = df['Gender'].value_counts()

# Mengambil daftar urutan kategori di sumbu X secara pasti
kategori_x = [tick.get_text() for tick in ax.get_xticklabels()]
jumlah_kategori = len(kategori_x)

# Menambahkan anotasi Angka & Persentase pada grafik
for i, p in enumerate(ax.patches):
    tinggi = p.get_height()
    if pd.isna(tinggi) or tinggi == 0:
        continue
        
    nama_kategori = kategori_x[i % jumlah_kategori]
    persentase = f'{(100 * tinggi / total_per_gender[nama_kategori]):.1f}%'
    
    ax.annotate(f'{int(tinggi)}\n({persentase})', 
                (p.get_x() + p.get_width() / 2., tinggi), 
                ha='center', va='bottom', 
                fontsize=11, fontweight='bold', color='black', xytext=(0, 5), 
                textcoords='offset points')

ax.set_ylim(0, max([p.get_height() for p in ax.patches if not pd.isna(p.get_height())]) * 1.15)
plt.tight_layout()
plt.show()

# ==========================================
# CETAK TABEL ANGKA DETAIL (PRINT OUT)
# ==========================================
print("\n" + "="*50)
print("📊 DETAIL ANGKA & PERSENTASE: GENDER VS CHURN")
print("="*50)

# Membuat tabulasi silang (Crosstab) untuk angka absolut dan persentase
ct_absolut = pd.crosstab(df['Gender'], df['Churn'])
ct_persen = pd.crosstab(df['Gender'], df['Churn'], normalize='index') * 100

for gender in ct_absolut.index:
    total = ct_absolut.loc[gender].sum()
    aman = ct_absolut.loc[gender, 0] if 0 in ct_absolut.columns else 0
    churn = ct_absolut.loc[gender, 1] if 1 in ct_absolut.columns else 0
    
    pct_aman = ct_persen.loc[gender, 0] if 0 in ct_persen.columns else 0.0
    pct_churn = ct_persen.loc[gender, 1] if 1 in ct_persen.columns else 0.0
    
    print(f"\n👤 [{gender}] - Total Populasi: {total} pelanggan")
    print(f"   ▶ Aman  (0) : {aman} pelanggan ({pct_aman:.1f}%)")
    print(f"   ▶ Churn (1) : {churn} pelanggan ({pct_churn:.1f}%)")
print("\n" + "="*50)
````

**Wawasan Analitik (Business Insight):**
* **Perbandingan Loyalitas:** Pelanggan Pria (*Male*) memiliki rasio *churn* yang sedikit lebih tinggi, yaitu sebesar **17,5% (532 pengguna)** dari total populasi pria. #### 4.3.1. Dampak Status Pernikahan terhadap Risiko Churn

Setelah mengetahui komposisi populasinya, pertanyaan kritis selanjutnya adalah: Apakah pelanggan yang masih lajang lebih mudah berpindah-pindah platform (*brand switching*) dibandingkan mereka yang sudah menetap dan berkeluarga? Kita akan membedahnya melalui visualisasi berikut.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Konfigurasi Tema dan Warna
sns.set_theme(style="whitegrid")
warna_1 = "#4cca96" # Mint Green (Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

print("Memproses visualisasi 4.3.1 (Marital Status vs Churn)...")

# ==========================================
# 4.3.1. Visualisasi Marital Status vs Churn
# ==========================================
fig, ax = plt.subplots(figsize=(8, 5))

# Mengunci urutan agar perhitungan persentase akurat
urutan_kategori = df['MaritalStatus'].value_counts().index
sns.countplot(data=df, x='MaritalStatus', hue='Churn', palette=warna_utama, order=urutan_kategori, ax=ax)

# Kustomisasi Grafik
ax.set_title('4.3.1 Proporsi Status Pernikahan Berdasarkan Status Churn', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Status Pernikahan', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)
ax.legend(title='Status', labels=['Aman (0)', 'Churn (1)'])

# Menghitung sub-total per kategori
total_per_status = df['MaritalStatus'].value_counts()
kategori_x = [tick.get_text() for tick in ax.get_xticklabels()]
jumlah_kategori = len(kategori_x)

# Menambahkan anotasi Angka & Persentase pada grafik
for i, p in enumerate(ax.patches):
    tinggi = p.get_height()
    if pd.isna(tinggi) or tinggi == 0:
        continue
        
    nama_kategori = kategori_x[i % jumlah_kategori]
    persentase = f'{(100 * tinggi / total_per_status[nama_kategori]):.1f}%'
    
    ax.annotate(f'{int(tinggi)}\n({persentase})', 
                (p.get_x() + p.get_width() / 2., tinggi), 
                ha='center', va='bottom', 
                fontsize=11, fontweight='bold', color='black', xytext=(0, 5), 
                textcoords='offset points')

ax.set_ylim(0, max([p.get_height() for p in ax.patches if not pd.isna(p.get_height())]) * 1.15)
plt.tight_layout()
plt.show()

# ==========================================
# CETAK TABEL ANGKA DETAIL (PRINT OUT)
# ==========================================
print("\n" + "="*50)
print("📊 DETAIL ANGKA & PERSENTASE: MARITAL STATUS VS CHURN")
print("="*50)

ct_absolut = pd.crosstab(df['MaritalStatus'], df['Churn'])
ct_persen = pd.crosstab(df['MaritalStatus'], df['Churn'], normalize='index') * 100

for status in urutan_kategori:
    total = ct_absolut.loc[status].sum()
    aman = ct_absolut.loc[status, 0] if 0 in ct_absolut.columns else 0
    churn = ct_absolut.loc[status, 1] if 1 in ct_absolut.columns else 0
    
    pct_aman = ct_persen.loc[status, 0] if 0 in ct_persen.columns else 0.0
    pct_churn = ct_persen.loc[status, 1] if 1 in ct_persen.columns else 0.0
    
    print(f"\n💍 [{status}] - Total Populasi: {total} pelanggan")
    print(f"   ▶ Aman  (0) : {aman} pelanggan ({pct_aman:.1f}%)")
    print(f"   ▶ Churn (1) : {churn} pelanggan ({pct_churn:.1f}%)")
print("\n" + "="*50)
````

**Wawasan Analitik (Business Insight):**
* **Perbandingan Loyalitas:** Terdapat disparitas risiko yang sangat mencolok berdasarkan status pernikahan. Pelanggan lajang (*Single*) memiliki tingkat *churn* yang paling tinggi dan mengkhawatirkan, yaitu mencapai **26,7% (415 pengguna)**. Sebaliknya, pelanggan yang sudah menikah (*Married*) terbukti sebagai kelompok paling loyal dengan rasio *churn* terendah di angka **11,3% (302 pengguna)**. Pelanggan bercerai (*Divorced*) berada di tengah dengan rasio *churn* sebesar **14,6% (124 pengguna)**.
* **Interpretasi Bisnis:** Pelanggan lajang cenderung lebih sensitif terhadap harga (*price-sensitive*) dan memiliki fleksibilitas tinggi untuk berpindah platform demi mengejar promosi terbaik (*brand switching*). Di sisi lain, pelanggan yang sudah berkeluarga lebih mengutamakan kenyamanan, keandalan pengiriman, dan rutinitas belanja bulanan, sehingga mereka enggan untuk berpindah E-Commerce. 
* **Tindakan Strategis:** Perusahaan harus merancang kampanye retensi yang secara spesifik menargetkan pelanggan lajang (misalnya program loyalitas individu, *flash sale* gawai, atau fesyen), sambil terus mempertahankan kualitas layanan operasional dasar untuk menjaga kenyamanan kelompok pelanggan yang sudah menikah.

````python
# Konfigurasi Tema dan Warna
sns.set_theme(style="whitegrid")
warna_1 = "#4cca96" # Mint Green (Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

print("Memproses visualisasi 4.6.1 (Number of Device vs Churn)...")

fig, ax = plt.subplots(figsize=(10, 6))

# Mengurutkan berdasarkan jumlah perangkat
order_kategori = sorted(df['NumberOfDeviceRegistered'].dropna().unique())

sns.countplot(
    data=df, 
    x='NumberOfDeviceRegistered', 
    hue='Churn', 
    palette=warna_utama, 
    order=order_kategori, 
    ax=ax
)

# Kustomisasi Grafik
ax.set_title('4.6.1 Proporsi Jumlah Perangkat Terdaftar Berdasarkan Status Churn', fontsize=15, fontweight='bold', pad=20)
ax.set_xlabel('Jumlah Perangkat (Unit)', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)
ax.legend(title='Status', labels=['Aman (0)', 'Churn (1)'])

# Menghitung sub-total per kategori untuk persentase rasio Churn
total_per_group = df['NumberOfDeviceRegistered'].value_counts()

# Menambahkan anotasi Angka & Persentase (Rasio per Grup)
for i, p in enumerate(ax.patches):
    tinggi = p.get_height()
    if pd.isna(tinggi) or tinggi == 0:
        continue
    
    # Mencari nama kategori berdasarkan urutan bar
    # i % len(order_kategori) memberikan index kategori yang tepat
    idx_kategori = i % len(order_kategori)
    nama_kategori = order_kategori[idx_kategori]
    
    # Hitung persentase terhadap total populasi di grup tersebut
    persentase = (tinggi / total_per_group[nama_kategori]) * 100
    
    ax.annotate(f'{int(tinggi)}\n({persentase:.1f}%)', 
                (p.get_x() + p.get_width() / 2., tinggi), 
                ha='center', va='bottom', 
                fontsize=10, fontweight='bold', color='black', 
                xytext=(0, 5), textcoords='offset points')

ax.set_ylim(0, max([p.get_height() for p in ax.patches if not pd.isna(p.get_height())]) * 1.2)
plt.tight_layout()
plt.show()

print("\n" + "="*65)
print("📊 DETAIL ANGKA & PERSENTASE: NUMBER OF DEVICE REGISTERED VS CHURN")
print("="*65)

ct_absolut = pd.crosstab(df['NumberOfDeviceRegistered'], df['Churn'])
ct_persen = pd.crosstab(df['NumberOfDeviceRegistered'], df['Churn'], normalize='index') * 100

for group in order_kategori:
    total = ct_absolut.loc[group].sum()
    aman = ct_absolut.loc[group, 0] if 0 in ct_absolut.columns else 0
    churn = ct_absolut.loc[group, 1] if 1 in ct_absolut.columns else 0
    
    pct_churn = ct_persen.loc[group, 1] if 1 in ct_persen.columns else 0.0
    
    print(f"▶ {int(group)} Perangkat : Total {total:<4} | Aman: {aman:<4} | Churn: {churn:<4} ({pct_churn:.1f}%)")
print("="*65 + "\n")
````

Wawasan Analitik (Business Insight):**

* **Tren Korelasi Positif:** Data menunjukkan pola yang sangat konsisten; **semakin banyak perangkat yang terdaftar, semakin tinggi risiko pelanggan untuk melakukan *churn***. Lonjakan risiko mulai terlihat signifikan saat pelanggan memiliki lebih dari 4 perangkat.
* **Titik Kritis (Red Flag):** Pelanggan dengan **6 perangkat** memiliki tingkat *churn* tertinggi mencapai **35,6%**. Angka ini hampir 4 kali lipat lebih tinggi dibandingkan pelanggan yang hanya memiliki 1-2 perangkat (8,6% - 8,8%).
* **Interpretasi Bisnis:** Tingginya *churn* pada pengguna banyak perangkat mengindikasikan dua hal:
    1.  **Penyalahgunaan Akun:** Akun mungkin digunakan secara kolektif (berbagi akun) oleh banyak individu demi mengejar promo atau diskon tertentu, sehingga tidak ada loyalitas personal terhadap platform.
    2.  **Isu Keamanan & Kenyamanan:** Semakin banyak perangkat yang tertaut, potensi kendala sinkronisasi atau kekhawatiran akan privasi data meningkat, yang dapat mendorong pelanggan untuk berhenti menggunakan layanan.
* **Tindakan Strategis:** Perusahaan disarankan untuk melakukan audit keamanan atau memberikan notifikasi verifikasi tambahan jika jumlah perangkat yang terdaftar melebihi batas wajar (misalnya >4 perangkat). ### 4.11. Riwayat Transaksi: Jumlah Total Pesanan (Order Count)

#### 4.11.1. Dampak Jumlah Total Pesanan terhadap Risiko Churn

Kenyataan bahwa 70% pelanggan tertahan di angka 1-2 pesanan menimbulkan pertanyaan penting: Apakah pelanggan yang jarang bertransaksi ini juga merupakan kelompok yang paling berisiko melakukan *churn*? Mari kita bandingkan probabilitas risikonya secara langsung.

````python
# Konfigurasi Tema dan Warna
sns.set_theme(style="whitegrid")
warna_1 = "#4cca96" # Mint Green (Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

print("Memproses visualisasi 4.11.1 (Order Count vs Churn)...")

fig, ax = plt.subplots(figsize=(14, 6))

# Mengurutkan berdasarkan jumlah pesanan
order_kategori = sorted(df['OrderCount'].dropna().unique())

sns.countplot(
    data=df, 
    x='OrderCount', 
    hue='Churn', 
    palette=warna_utama, 
    order=order_kategori, 
    ax=ax
)

# Kustomisasi Grafik
ax.set_title('4.11.1 Proporsi Jumlah Total Pesanan Berdasarkan Status Churn', fontsize=15, fontweight='bold', pad=20)
ax.set_xlabel('Total Pesanan (Kali Transaksi)', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)
ax.legend(title='Status', labels=['Aman (0)', 'Churn (1)'])

# Menghitung sub-total per kategori untuk persentase rasio Churn
total_per_group = df['OrderCount'].value_counts()

# Mengambil label sumbu X secara langsung dari grafik untuk memetakan kategori dengan presisi mutlak
kategori_x_labels = [float(tick.get_text()) for tick in ax.get_xticklabels()]

# Menambahkan anotasi Angka & Persentase
for p in ax.patches:
    tinggi = p.get_height()
    if pd.isna(tinggi) or tinggi == 0:
        continue
    
    # Mendapatkan koordinat tengah dari batang grafik (bar)
    x_pos = p.get_x() + p.get_width() / 2.0
    idx_kategori = int(round(x_pos))
    
    # Mencocokkan nama kategori berdasarkan indeks sumbu X
    nama_kategori = kategori_x_labels[idx_kategori]
    
    # Menghitung persentase terhadap total populasi di kelompok kategori tersebut
    persentase = (tinggi / total_per_group[nama_kategori]) * 100
    
    # Ukuran font disesuaikan agar tidak tumpang tindih karena jumlah bar yang banyak
    ax.annotate(f'{int(tinggi)}\n({persentase:.1f}%)', 
                (p.get_x() + p.get_width() / 2., tinggi), 
                ha='center', va='bottom', 
                fontsize=8, fontweight='bold', color='black', 
                xytext=(0, 5), textcoords='offset points')

ax.set_ylim(0, max([p.get_height() for p in ax.patches if not pd.isna(p.get_height())]) * 1.2)
plt.tight_layout()
plt.show()

print("\n" + "="*75)
print("📊 DETAIL ANGKA & PERSENTASE: ORDER COUNT VS CHURN")
print("="*75)

ct_absolut = pd.crosstab(df['OrderCount'], df['Churn'])
ct_persen = pd.crosstab(df['OrderCount'], df['Churn'], normalize='index') * 100

for order in order_kategori:
    total = ct_absolut.loc[order].sum()
    aman = ct_absolut.loc[order, 0] if 0 in ct_absolut.columns else 0
    churn = ct_absolut.loc[order, 1] if 1 in ct_absolut.columns else 0
    
    pct_churn = ct_persen.loc[order, 1] if 1 in ct_persen.columns else 0.0
    
    print(f"▶ {int(order):<2} Kali Pesanan : Total {total:<4} | Aman: {aman:<4} | Churn: {churn:<4} ({pct_churn:.1f}%)")
print("="*75 + "\n")
````

**Wawasan Analitik (Business Insight):**

* **Ambang Batas Loyalitas (Loyalty Threshold):** Data menunjukkan sebuah pola transisi yang sangat jelas. Fase kritis retensi berada pada 1 hingga 3 kali pesanan, di mana tingkat *churn* cukup tinggi (16,6% - 18,8%). Namun, begitu pelanggan berhasil dibujuk untuk melakukan **pesanan ke-4 dan ke-5**, tingkat *churn* menurun drastis hingga mencapai titik terendah di angka **8,0%**.
* **Interpretasi Bisnis:** Pesanan ke-4 dan ke-5 merupakan "Fase Pembentukan Kebiasaan" (*Habit-Forming Phase*). Pelanggan yang telah mencapai titik ini sudah sepenuhnya familier dengan antarmuka aplikasi, puas dengan kualitas logistik, dan mengintegrasikan platform ke dalam rutinitas belanja mereka. Lonjakan risiko pada angka pesanan yang sangat tinggi (di atas 7 kali) lebih mencerminkan volatilitas statistik karena ukuran sampel populasi yang terlampau kecil, atau indikasi bahwa siklus hidup pelanggan (*customer lifecycle*) secara natural telah berakhir (kebutuhan mereka telah terpenuhi sepenuhnya).

* **Karakteristik Pemburu Diskon (Promo-Hunter):** Terdapat anomali di mana tingkat *churn* justru meningkat secara konstan ketika penggunaan kupon melebihi 6 lembar, mencapai puncaknya pada kelompok **8 kupon (21,4%)**. Begitu alokasi kupon mereka habis, mereka sangat rentan untuk berpindah mencari platform lain yang sedang melakukan promosi besar-besaran. 
* **Tindakan Strategis:** Perusahaan direkomendasikan untuk menerapkan kebijakan "Pembatasan Kupon Pintar" (*Smart Coupon Capping*). Alih-alih memberikan kupon tak terbatas, distribusikan promosi secara terukur dengan batas ideal 3 lembar per pengguna aktif dalam periode tertentu. Hal ini terbukti memberikan tingkat loyalitas terbaik sekaligus menjaga margin keuntungan (*profit margin*) perusahaan dari kebocoran anggaran promosi.

````python
# Konfigurasi Tema dan Warna
sns.set_theme(style="whitegrid")
warna_1 = "#4cca96" # Mint Green (Aman)
warna_2 = "#72c8cb" # Soft Cyan (Churn)
warna_utama = [warna_1, warna_2]

print("Memproses visualisasi 4.15.1 (Day Since Last Order vs Churn)...")

fig, ax = plt.subplots(figsize=(16, 6))

# Mengurutkan berdasarkan hari
order_kategori = sorted(df['DaySinceLastOrder'].dropna().unique())

sns.countplot(
    data=df, 
    x='DaySinceLastOrder', 
    hue='Churn', 
    palette=warna_utama, 
    order=order_kategori, 
    ax=ax
)

# Kustomisasi Grafik
ax.set_title('4.15.1 Proporsi Hari Sejak Pesanan Terakhir Berdasarkan Status Churn', fontsize=15, fontweight='bold', pad=20)
ax.set_xlabel('Jeda Waktu (Hari)', fontsize=12)
ax.set_ylabel('Jumlah Pelanggan', fontsize=12)
ax.legend(title='Status', labels=['Aman (0)', 'Churn (1)'])

# Menghitung sub-total per kategori untuk persentase rasio Churn
total_per_group = df['DaySinceLastOrder'].value_counts()

# Mengambil label sumbu X secara langsung dari grafik untuk memetakan kategori
kategori_x_labels = [float(tick.get_text()) for tick in ax.get_xticklabels()]

# Menambahkan anotasi Angka & Persentase
for p in ax.patches:
    tinggi = p.get_height()
    if pd.isna(tinggi) or tinggi == 0:
        continue
    
    # Mendapatkan koordinat tengah dari batang grafik
    x_pos = p.get_x() + p.get_width() / 2.0
    idx_kategori = int(round(x_pos))
    
    # Mencocokkan nama kategori berdasarkan indeks sumbu X
    nama_kategori = kategori_x_labels[idx_kategori]
    
    # Menghitung persentase terhadap total populasi di kelompok kategori tersebut
    persentase = (tinggi / total_per_group[nama_kategori]) * 100
    
    ax.annotate(f'{int(tinggi)}\n({persentase:.1f}%)', 
                (p.get_x() + p.get_width() / 2., tinggi), 
                ha='center', va='bottom', 
                fontsize=8, fontweight='bold', color='black', 
                xytext=(0, 5), textcoords='offset points')

ax.set_ylim(0, max([p.get_height() for p in ax.patches if not pd.isna(p.get_height())]) * 1.25)
plt.tight_layout()
plt.show()

print("\n" + "="*75)
print("📊 DETAIL ANGKA & PERSENTASE: DAY SINCE LAST ORDER VS CHURN")
print("="*75)

ct_absolut = pd.crosstab(df['DaySinceLastOrder'], df['Churn'])
ct_persen = pd.crosstab(df['DaySinceLastOrder'], df['Churn'], normalize='index') * 100

for day in order_kategori:
    total = ct_absolut.loc[day].sum()
    aman = ct_absolut.loc[day, 0] if 0 in ct_absolut.columns else 0
    churn = ct_absolut.loc[day, 1] if 1 in ct_absolut.columns else 0
    
    pct_churn = ct_persen.loc[day, 1] if 1 in ct_persen.columns else 0.0
    
    print(f"▶ {int(day):<2} Hari : Total {total:<4} | Aman: {aman:<4} | Churn: {churn:<4} ({pct_churn:.1f}%)")
print("="*75 + "\n")
````


**Wawasan Analitik (Business Insight):**

* **Titik Stabilitas (*Stability Point*):** Begitu pelanggan melewati jendela krisis 48 jam pertama, tingkat *churn* langsung anjlok ke angka **15,5%** pada hari ke-2, dan terus menunjukkan tren stabilitas hingga berangsur menurun seiring lamanya jeda waktu.
* **Tindakan Strategis:** Perusahaan perlu mengevaluasi kembali skema pemasaran agar tidak mengundang kelompok *fraudulent* (penyalahguna promo). Fokus bisnis harus dikembalikan pada pembangunan (integritas) dan kepercayaan jangka panjang. Alih-alih memberikan seluruh diskon di awal secara tunai, perusahaan dapat mengonversi promo tersebut menjadi *loyalty points* yang baru dapat dicairkan pada transaksi berikutnya, guna mencegah kerugian dari perilaku *hit-and-run*.

    2. **Kepuasan Transaksional (Hit-and-Run):** Di sisi lain, tingginya *churn* pada Skor 5 kembali mengonfirmasi fenomena pemburu promo instan. Mereka memberikan bintang lima semata-mata karena mereka sangat puas mendapatkan barang murah bersubsidi besar, namun mereka tidak memiliki intensi untuk membangun loyalitas jangka panjang. Begitu transaksi menguntungkan tersebut selesai, mereka segera meninggalkan platform.
* **Tindakan Strategis:** Perusahaan harus berhenti menggunakan *Customer Satisfaction Score* (CSAT) sebagai metrik utama untuk memprediksi retensi, karena terbukti sangat bias. Mulailah mengadopsi metrik *Net Promoter Score* (NPS) yang lebih akurat dalam mengukur loyalitas aktual. Selain itu, berikan apresiasi (seperti diskon kompensasi) kepada pelanggan setia di kelompok Skor 1 dan 2 sebagai bentuk permohonan maaf dan penghargaan atas umpan balik mereka yang berharga.

2. **Isolasi Variabel Prediktor (`X`):** Fitur-fitur operasional dipisahkan dari variabel target. Kumpulan fitur ini merepresentasikan profil empiris pelanggan (seperti usia berlangganan, riwayat keluhan, dan sensitivitas terhadap promosi) yang akan dianalisis oleh algoritma untuk menemukan pola risiko *churn*.

### 5.2 Pemisahan Data Latih dan Data Uji (Train-Test Split)

Langkah berikutnya sebelum melakukan transformasi apa pun adalah membagi dataset menjadi dua bagian yang terisolasi: Data Latih (*Training Set*) dan Data Uji (*Testing Set*). Proses ini sangat penting bagi pihak manajemen karena berfungsi sebagai simulasi dunia nyata. Data latih digunakan oleh algoritma untuk mempelajari pola perilaku pelanggan, sedangkan data uji disembunyikan dan baru digunakan di tahap akhir untuk menguji performa prediksi model pada data yang belum pernah dilihat sebelumnya.

Dalam konteks dataset analitik ini, terdapat satu tantangan utama, yaitu ketidakseimbangan proporsi pelanggan yang *churn* (berkisar 16,6%) dan pelanggan yang loyal (berkisar 83,4%). Oleh karena itu, penggunaan parameter stratifikasi (`stratify`) menjadi suatu kewajiban. Parameter ini menjamin bahwa rasio pelanggan berisiko akan tetap konsisten, baik pada fase pembelajaran maupun fase pengujian, sehingga metrik evaluasi yang dilaporkan kepada tim bisnis nantinya bersifat objektif dan bebas dari bias statistik.

### 5.3 Integrated Feature Transformation

Algoritma *Machine Learning* hanya dapat memproses instruksi dalam format numerik matematis, sehingga data metrik dan profil pelanggan harus dikonversi. Bagi pihak manajemen dan tim teknis, fase ini sangat kritikal untuk memastikan keandalan sistem saat kelak diimplementasikan di dunia nyata (*production environment*).

Pendekatan statis (seperti pengubahan teks ke angka secara manual tanpa memori penyimpanan) sangat berisiko menyebabkan kerusakan sistem (*system failure*) di masa depan, terutama apabila sistem menerima data pelanggan baru dengan kategori yang belum pernah ada sebelumnya. Oleh karena itu, arsitektur transformasi pada tahap ini dirancang menggunakan cetak biru terintegrasi (`ColumnTransformer`) yang mencakup dua standardisasi utama:

1. **Pengodean Kategorikal (Categorical Encoding):** Mengonversi atribut profil (seperti jenis perangkat atau metode pembayaran) menjadi matriks biner. Konfigurasi `handle_unknown='ignore'` diterapkan secara spesifik agar sistem tetap stabil dan tidak mengalami *error* apabila kelak menghadapi variasi data pelanggan baru.
2. **Penyesuaian Skala Numerik (Numerical Scaling):** Menyeragamkan rentang indikator operasional (seperti nominal *Cashback* dan jarak logistik). Pemilihan algoritma `RobustScaler` diinstruksikan karena kemampuannya dalam meredam bias dari nilai-nilai ekstrem (*outliers*), sehingga algoritma prediksi nantinya tidak hanya berfokus pada pelanggan dengan transaksi yang tidak wajar.

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler

# 1. Identifikasi dinamis untuk memisahkan tipe data secara otomatis
kolom_kategorikal = X_train.select_dtypes(include=['object']).columns.tolist()
kolom_numerik = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# 2. Menampilkan ringkasan klasifikasi kolom
print("--- Ringkasan Taksonomi Fitur ---")
print(f"Jumlah Fitur Kategorikal : {len(kolom_kategorikal)} kolom")
print(f"Jumlah Fitur Numerik     : {len(kolom_numerik)} kolom\n")

# 3. Konstruksi cetak biru transformasi (Preprocessor Blueprint)
preprocessor = ColumnTransformer(
    transformers=[
        # Transformasi teks: drop='first' mencegah multikolinearitas, handle_unknown menjaga stabilitas sistem
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), kolom_kategorikal),
        # Transformasi angka: RobustScaler meredam anomali metrik bisnis yang fluktuatif
        ('num', RobustScaler(), kolom_numerik)
    ],
    # Membiarkan kolom di luar kategori di atas (jika ada) tetap utuh
    remainder='passthrough'
)

# 4. Membuktikan bahwa arsitektur ColumnTransformer telah terbentuk
print("--- Bukti Konstruksi Preprocessor ---")
preprocessor
````

**Penjelasan Hasil Eksekusi (Output):**
Berdasarkan log sistem, proses identifikasi telah memisahkan 5 fitur sebagai data kategorikal dan 13 fitur sebagai data numerik secara otomatis. 

Keluaran visual (bagan interaktif) membuktikan bahwa arsitektur `ColumnTransformer` telah berhasil dikonstruksi. Bagan tersebut memperlihatkan alur pemrosesan di mana kelompok fitur teks (`cat`) diarahkan ke modul `OneHotEncoder`, sementara kelompok fitur angka (`num`) diarahkan ke modul `RobustScaler`. 

Ketiadaan transformasi fisik pada tahap ini merupakan indikator positif yang menjamin bahwa proyek ini terbebas dari kebocoran informasi (*Data Leakage*). Matriks data pelanggan saat ini tetap dalam kondisi aslinya, dan proses standardisasi ini telah terkunci dalam sebuah arsitektur yang baru akan dieksekusi secara otomatis di dalam siklus algoritma tertutup pada tahap evaluasi model berikutnya.

---
**Key Points (Poin Utama 5.3):**
* **Stabilitas Masa Depan:** Transformasi berbasis memori (dinamis) mencegah kegagalan sistem akibat kategori pelanggan baru.
* **Ketahanan terhadap Anomali:** Penggunaan *RobustScaler* memastikan performa algoritma tidak terdistorsi oleh lonjakan transaksi yang ekstrem.
* **Bebas Kebocoran Data:** Transformasi direkam sebagai cetak biru arsitektur, tidak ada perubahan struktural fisik yang mencemari data uji.

#### 5.4 Konstruksi Pipeline Komparatif dan Strategi Resampling (Comparative Pipeline Construction)

Kondisi populasi pelanggan pada dataset ini mengalami ketimpangan (*imbalanced data*), di mana kelompok pelanggan berisiko (*churn*) hanya merepresentasikan sekitar 16,6% dari total populasi. Jika algoritma dilatih pada kondisi alami ini, model cenderung akan mendominasi prediksi pada kelas mayoritas (pelanggan loyal) demi mencapai tingkat akurasi umum yang tinggi, namun gagal mendeteksi pelanggan yang sebenarnya akan pergi (menghasilkan *False Negative* yang memicu kerugian finansial tinggi).

Untuk mengatasi hal tersebut, intervensi penyeimbangan data menggunakan metode SMOTE (*Synthetic Minority Over-sampling Technique*) diperlukan. Namun, bagi pihak manajemen dan tim evaluasi analitik, efektivitas intervensi ini tidak boleh sekadar diasumsikan; melainkan harus dibuktikan secara lengkap dan detail. Oleh karena itu, kita merancang dua arsitektur *Pipeline* komparatif:
1. **Pipeline Baseline (Tanpa Resampling):** Berfungsi sebagai kelompok kontrol (kondisi alami).
2. **Pipeline SMOTE (Dengan Resampling):** Berfungsi sebagai kelompok eksperimen.

**Pencegahan Kebocoran Data (Data Leakage):** Konstruksi *Pipeline* ini secara khusus menggunakan pustaka `imblearn`. Hal ini merupakan protokol keamanan mutlak. Jika SMOTE diaplikasikan secara manual ke seluruh data latih sebelum proses validasi silang (*Cross-Validation*), data sintetis akan mencemari area validasi, menghasilkan laporan performa yang menyesatkan (*misinformation*). Dengan arsitektur `imblearn Pipeline`, sintesis data baru akan ditembakkan secara dinamis *hanya* pada saat algoritma belajar, menjaga area pengujian tetap murni sebagai representasi dunia nyata.

````python
X_train.dtypes
````

````python
# PENTING: Menggunakan Pipeline dari imblearn, BUKAN dari sklearn
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# 1. SKENARIO 1: Arsitektur Kontrol (Tanpa Intervensi Resampling)
# Arsitektur ini hanya memuat standardisasi data (preprocessor dari Sub-bab 5.3)
pipeline_no_smote = ImbPipeline(steps=[
    ('preprocessor', preprocessor)
    # Catatan: Slot untuk model Machine Learning akan ditambahkan pada Bab 6
])

# 2. SKENARIO 2: Arsitektur Eksperimen (Dengan Intervensi SMOTE)
# Arsitektur ini memuat standardisasi data dilanjutkan dengan sintesis kelas minoritas
# Parameter random_state=42 dikunci untuk memastikan hasil eksperimen dapat direplikasi
pipeline_smote = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42))
    # Catatan: Slot untuk model Machine Learning akan ditambahkan pada Bab 6
])

# 3. Validasi status konstruksi
print("--- Status Konstruksi Arsitektur Pipeline ---")
print("1. [BERHASIL] pipeline_no_smote : Siap untuk pengujian Baseline.")
print("2. [BERHASIL] pipeline_smote    : Siap untuk pengujian penanganan Imbalanced Data.")
print("\nKedua arsitektur di atas telah mengunci 'preprocessor' dan aman dari risiko Data Leakage.")
````

Pada tahap ini, sangat penting untuk ditekankan bahwa **tidak ada satu pun baris data sintetis yang dihasilkan atau ditambahkan ke dalam `X_train`**. #### 6.3 Model Initialization and Benchmarking (Cross-Validation)

**Penjelasan Langkah dan Signifikansi Bisnis:**
Untuk menemukan model prediktif yang terbaik dalam mendeteksi probabilitas *churn*, tim analitik tidak boleh bergantung pada asumsi algoritma tunggal. Oleh karena itu, kita menginisialisasi 5 (lima) algoritma klasifikasi yang mewakili hierarki kompleksitas matematis yang berbeda:
1. **Logistic Regression:** Algoritma linear transparan yang berfungsi sebagai batas bawah (*baseline*) performa.
2. **Decision Tree:** Algoritma berbasis aturan bercabang (*rule-based*) yang mudah divalidasi logikanya oleh manajemen.
3. **Random Forest:** Algoritma *Ensemble Bagging* yang menggabungkan banyak *Decision Tree* untuk menekan varians dan mencegah model terlalu menghafal data (*overfitting*).
4. **Gradient Boosting:** Algoritma *Ensemble Boosting* yang bekerja secara berantai, di mana setiap iterasi secara khusus memperbaiki kesalahan prediksi dari iterasi sebelumnya.
5. **XGBoost:** Algoritma mutakhir berskala industri yang sangat optimal dalam menangani data tabular kompleks dan memiliki sistem penalti internal untuk menjaga keseimbangan model.

Kelima algoritma ini akan diuji menggunakan metode **Stratified 5-Fold Cross-Validation**. Data latih akan dibelah menjadi 5 bagian, dan setiap model akan diuji berulang kali pada skenario lipatan yang berbeda. Pengujian dilakukan secara paralel pada dua arsitektur *Pipeline* (Alami vs SMOTE) yang hasilnya akan disajikan dalam tabel terpisah. Selain melihat rata-rata kemampuan deteksi (*Avg Recall*), kita juga menghitung standar deviasi (*Std Recall*) untuk memvalidasi stabilitas performa model di berbagai skenario data.

````python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# Mengimpor 5 algoritma kandidat yang valid dan tangguh terhadap anomali sintetik
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# 1. Inisialisasi Dictionary Model
# Parameter random_state dikunci pada nilai 42 agar hasil pengujian selalu stabil dan dapat direplikasi (reproducible)
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, eval_metric='logloss')
}

# 2. Pengaturan Cross Validation (5 Lipatan)
# Menggunakan stratifikasi untuk memastikan rasio kelas (Churn vs Loyal) tetap seimbang di setiap lipatan
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Array terpisah untuk menampung hasil evaluasi
hasil_alami = []
hasil_smote = []

print("Memulai proses benchmarking (Stratified 5-Fold Cross-Validation)...")
print("Sistem mengevaluasi 5 model pada Pipeline Baseline & SMOTE. Harap tunggu...\n")

# 3. Looping Evaluasi Model Terintegrasi
for name, model in models.items():
    print(f"-> Memproses algoritma: {name}")
    
    # Skenario A: Pipeline Baseline (Kondisi Alami tanpa Resampling)
    pipe_baseline = ImbPipeline(steps=[
        ('preprocessor', preprocessor), # Mengambil cetak biru preprocessor dari Bab 5
        ('classifier', model)
    ])
    
    # Skenario B: Pipeline Eksperimen (Dengan Intervensi SMOTE)
    pipe_smote = ImbPipeline(steps=[
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])
    
    # Menghitung skor Recall secara objektif (n_jobs=-1 menggunakan seluruh inti CPU untuk efisiensi waktu)
    cv_recall_baseline = cross_val_score(pipe_baseline, X_train, y_train, cv=skf, scoring='recall', n_jobs=-1)
    cv_recall_smote = cross_val_score(pipe_smote, X_train, y_train, cv=skf, scoring='recall', n_jobs=-1)
    
    # Mencatat rata-rata dan standar deviasi performa (Kondisi Alami)
    hasil_alami.append({
        'Algoritma': name,
        'Avg Recall': np.mean(cv_recall_baseline),
        'Std Recall': np.std(cv_recall_baseline)
    })
    
    # Mencatat rata-rata dan standar deviasi performa (Dengan SMOTE)
    hasil_smote.append({
        'Algoritma': name,
        'Avg Recall': np.mean(cv_recall_smote),
        'Std Recall': np.std(cv_recall_smote)
    })

# 4. Merangkum hasil ke dalam struktur tabel DataFrame dan mengurutkan dari nilai terbesar ke terkecil
df_alami = pd.DataFrame(hasil_alami).sort_values(by='Avg Recall', ascending=False).reset_index(drop=True)
df_smote = pd.DataFrame(hasil_smote).sort_values(by='Avg Recall', ascending=False).reset_index(drop=True)

print("\n--- Proses Selesai ---")

# Menampilkan Tabel 1 (Alami)
print("\nTabel 1: Performa Model pada Kondisi Alami (Tanpa Resampling)")
display(df_alami.style.format({
    'Avg Recall': "{:.4f}",
    'Std Recall': "{:.4f}"
}).background_gradient(cmap='Oranges', subset=['Avg Recall']))

# Menampilkan Tabel 2 (SMOTE)
print("\nTabel 2: Performa Model dengan Intervensi SMOTE")
display(df_smote.style.format({
    'Avg Recall': "{:.4f}",
    'Std Recall': "{:.4f}"
}).background_gradient(cmap='GnBu', subset=['Avg Recall']))
````

**Penjelasan Hasil Eksekusi (Output):**
Kedua tabel komparasi di atas menyajikan rata-rata performa metrik prediksi utama (*Recall*) beserta tingkat stabilitasnya (*Std Recall*) dari 5 algoritma berdasarkan pengujian *Cross-Validation* 5 lipatan. Hasil ini memberikan landasan empiris yang sangat terukur untuk pemilihan model final:

   - **Wawasan Bisnis:** Algoritma pohon tunggal (*single tree*) sangat rentan terhadap *overfitting* saat dihadapkan pada data sintetik (titik data baru hasil interpolasi) yang berada tepat di perbatasan keputusan. Meskipun arsitektur XGBoost dengan parameter bawaan (*default*) telah memenangkan tahap *benchmarking*, tim analitik memiliki kewajiban untuk memastikan bahwa model ini beroperasi pada kapasitas maksimalnya. Setiap algoritma memiliki konfigurasi internal yang mengontrol bagaimana ia belajar—seperti seberapa dalam pohon keputusan boleh tumbuh, atau seberapa cepat ia mengoreksi kesalahan.

Pada tahap ini, kita melakukan *Hyperparameter Tuning* menggunakan metode `RandomizedSearchCV`. Metode ini akan menguji berbagai kombinasi konfigurasi secara acak yang terukur pada area Data Latih melalui mekanisme *Cross-Validation*. Tujuan tunggal dari proses ini adalah memeras setiap potensi yang ada untuk meningkatkan metrik **Recall**, memastikan bahwa sistem dapat menangkap sinyal *churn* yang lebih halus sebelum arsitektur ini dikunci dan dihadapkan pada simulasi dunia nyata (Data Uji).

````python
from sklearn.model_selection import RandomizedSearchCV

print("Memulai proses Hyperparameter Tuning untuk XGBoost + SMOTE...")

# 1. Mendefinisikan ulang arsitektur pemenang secara spesifik
xgb_model = XGBClassifier(random_state=42, eval_metric='logloss')
pipe_final = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', xgb_model)
])

# 2. Mendefinisikan ruang pencarian parameter (Hyperparameter Grid)
# Awalan 'classifier__' wajib digunakan karena XGBoost berada di dalam objek Pipeline
param_distributions = {
    'classifier__max_depth': [3, 5, 7, 9],               # Kedalaman maksimal pohon (mencegah overfitting)
    'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2], # Kecepatan algoritma belajar dari kesalahan
    'classifier__n_estimators': [100, 200, 300, 500],    # Jumlah total pohon keputusan yang dibangun
    'classifier__subsample': [0.6, 0.8, 1.0],            # Persentase sampel acak untuk tiap pohon
    'classifier__colsample_bytree': [0.6, 0.8, 1.0]      # Persentase fitur acak untuk tiap pohon
}

# 3. Mengonfigurasi mekanisme pencarian dengan RandomizedSearchCV
# n_iter=20 berarti sistem akan menguji 20 kombinasi acak terbaik untuk menghemat waktu komputasi
random_search = RandomizedSearchCV(
    estimator=pipe_final,
    param_distributions=param_distributions,
    n_iter=20,
    scoring='recall',
    cv=skf,             # Menggunakan Stratified 5-Fold dari sub-bab sebelumnya
    random_state=42,
    n_jobs=-1,          # Menggunakan seluruh core CPU
    verbose=1           # Menampilkan log proses
)

# 4. Mengeksekusi pencarian pada Data Latih
random_search.fit(X_train, y_train)

# 5. Menampilkan hasil optimasi
print("\n--- Hasil Hyperparameter Tuning ---")
print("Kombinasi Parameter Terbaik:")
for param, value in random_search.best_params_.items():
    print(f"- {param.replace('classifier__', '')}: {value}")

print(f"\nSkor Recall XGBoost (Default) : {df_smote.loc[0, 'Avg Recall']:.4f}")
print(f"Skor Recall XGBoost (Tuned)   : {random_search.best_score_:.4f}")

# Menyimpan model yang sudah dioptimasi untuk digunakan pada tahap Test Set
best_model = random_search.best_estimator_
````

**Penjelasan Hasil Eksekusi (Output):**
Proses komputasi telah berhasil mengevaluasi 20 kombinasi arsitektur internal dari algoritma XGBoost. Berdasarkan hasil iterasi, kombinasi terbaik yang ditemukan mencatat metrik *Recall* sebesar **0.8292**, sedikit di bawah performa versi bawaan pabrik (*Default*) yang mencatat angka **0.8366**.

Fenomena analitik ini sangat wajar dan memberikan dua wawasan penting bagi tim bisnis:

Model telah beroperasi pada kapasitas maksimalnya sejak awal tanpa terjebak pada kompleksitas yang tidak perlu.

**Keputusan Strategis:** memutuskan untuk mendiskualifikasi hasil *Hyperparameter Tuning* dan secara resmi menggunakan **XGBoost (Default) + SMOTE** sebagai arsitektur final untuk pengujian Data Uji.

Namun, berdasarkan prinsip *Bias-Variance Tradeoff* dalam *Data Science*, fenomena ini dapat dipertanggungjawabkan sebagai **Overfitting Rasional (Benign Overfitting)** yang sangat aman, dengan landasan argumentasi berikut:

1. **Efek Kepadatan Data Sintetik (SMOTE):** Injeksi SMOTE menciptakan ribuan titik data buatan yang membuat kelas minoritas menjadi sangat padat. Hal ini mempermudah algoritma untuk menarik batas keputusan yang sempurna (100%) pada saat fase pelatihan.

SMOTE (*Synthetic Minority Over-sampling Technique*) bekerja dengan menciptakan observasi sintetis pelanggan *churn* berdasarkan interpolasi *k-nearest neighbor*. Dengan lebih banyak contoh *churn* (sintetis maupun asli), algoritma dapat **memperluas boundary decision** yang lebih akurat dan less biased terhadap kelas mayoritas. Peningkatan Recall yang moderat (+0.0134 atau 1.63%) bersifat modest namun signifikan secara statistik dan bisnis, membuktikan bahwa XGBoost sudah cukup robust pada kondisi alami, dan SMOTE berfungsi sebagai *fine-tuning* yang mengasah kepekaan model lebih jauh.

### 7.4 Limitasi Pemodelan dan Saran Pengembangan Lanjutan

Sebuah sistem *Machine Learning*, seakurat apa pun metrik evaluasinya pada saat pengujian, tetaplah representasi matematis dari data masa lalu. Untuk memastikan adopsi teknologi yang aman dan berkelanjutan oleh manajemen, batasan-batasan sistem ini harus dipetakan secara transparan, diikuti dengan peta jalan untuk pengembangannya di masa depan.

#### A. Limitasi Pemodelan (Modeling Limitations)
1. **Pergeseran Perilaku Konsumen (Concept Drift):** Model XGBoost ini dilatih menggunakan cuplikan data historis pada satu periode waktu tertentu. Perilaku konsumen *e-commerce* sangat dinamis; perubahan tren pasar, peluncuran fitur baru di dalam aplikasi, atau perubahan kebijakan harga dapat membuat pola historis yang dipelajari model menjadi usang seiring berjalannya waktu.
2. **Keterbatasan Variabel Eksternal:** Sistem peringatan dini ini murni beroperasi secara terisolasi menggunakan data interaksi internal (*in-app data*). Faktor-faktor eksternal yang sangat memengaruhi *churn*—seperti agresivitas promosi dari kompetitor, kondisi makroekonomi, atau sentimen viral di media sosial—tidak dapat ditangkap oleh algoritma saat ini.
3. **Ketiadaan Dimensi Runtun Waktu (Time-Series Blindness):** Format dataset saat ini bersifat agregat statis (akumulasi). Sebagai contoh, algoritma mengetahui "Total Jam Penggunaan Aplikasi" (*Tenure*), tetapi algoritma tidak mengetahui apakah durasi penggunaan tersebut menunjukkan tren peningkatan atau justru penurunan tajam dalam tiga bulan terakhir.

#### B. Saran Pengembangan Lanjutan (Future Development)
Untuk menjembatani batasan-batasan di atas dan meningkatkan nilai komersial dari sistem analitik ini, direkomendasikan tiga inisiatif strategis untuk fase pengembangan berikutnya:

1. **Integrasi Metrik RFM (Recency, Frequency, Monetary):** Pada iterasi pemodelan berikutnya, tim *Data Engineering* disarankan untuk mengekstraksi metrik *Recency* (kapan terakhir kali pelanggan bertransaksi), *Frequency* (seberapa sering dalam sebulan), dan *Monetary* (total nilai uang yang dihabiskan). Memasukkan indikator deret waktu ini ke dalam arsitektur XGBoost akan memberikan konteks yang jauh lebih tajam mengenai lintasan penurunan (*trajectory*) seorang pelanggan sebelum mereka benar-benar melakukan *churn*.
2. **Otomatisasi Pelatihan Ulang (MLOps / Continuous Training):** Untuk mencegah degradasi performa (*model decay*) akibat *concept drift*, arsitektur pemodelan ini harus diintegrasikan dengan *pipeline Machine Learning Operations* (MLOps). Sistem harus dirancang untuk secara otomatis melatih ulang dirinya sendiri secara periodik (misalnya setiap kuartal) setiap kali data pelanggan kuartal terbaru masuk ke dalam basis data perusahaan.
3. **Pelaksanaan A/B Testing untuk Validasi Strategi:** Sebelum manajemen meluncurkan kampanye retensi massal (berdasarkan rekomendasi di Sub-bab 7.2), disarankan untuk melakukan eksperimen *A/B Testing*. Sistem XGBoost dapat digunakan untuk mengidentifikasi 1.000 pelanggan berisiko tinggi. Kelompok ini kemudian dibagi dua: Kelompok A diberikan insentif retensi (misal: *cashback*), dan Kelompok B tidak diberikan apa-apa. Eksperimen ini krusial untuk mengukur secara faktual *Return on Investment* (ROI) dari rekomendasi analitik terhadap pencegahan *churn* di dunia nyata.
